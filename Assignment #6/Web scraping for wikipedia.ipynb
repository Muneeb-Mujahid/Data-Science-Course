```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get("https://en.wikipedia.org/wiki/Web_scraping")
page.status_code
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
title = soup.find("h1", id="firstHeading")
print(title.text)
```

```python
data = soup.find("div", id="mw-content-text")
print(data)
```

```python
data = pd.read_csv("data.csv")
data
```

```python
for i in data['Names']:
    data.replace(" ","_")
    print(i)
```

```python
data['Names']
```

```python
data['Names'].replace(to_replace=' ', value='_', inplace=True)
data["Names"]
```

```python
data['Names changed'] = data['Names'].str.replace(" ","_")
```

```python
data["Names"]
data['Names changed']
```

```python
def change(val):
    val2 = val.split(" ")
    val2 = "_".join(val2)
    print(val2)
    return val2
```

```python
data["Names New"] = data['Names'].apply(change)
```

```python
data.head()
```

```python
names = list(data["Names changed"])
names
```

```python
titles = []
data2 = []
#datas = []

for i in names:
    page = requests.get(f"https://en.wikipedia.org/wiki/{i}")
    soup = BeautifulSoup(page.content,'html.parser')
    title = soup.find("h1", id="firstHeading").text
    content = soup.find("div", id="mw-content-text").text
   # content1 = soup.find('span', class_:'mw-headline').text
    titles.append(title)
    data2.append(content)
   # datas.append(content1)
print(titles)
```

```python
all_data = {"Names":titles, "h1":data2}
all_data
```

```python
df = pd.DataFrame(all_data)
df.to_csv("scrapped_data(h1).csv",index=False)
```
